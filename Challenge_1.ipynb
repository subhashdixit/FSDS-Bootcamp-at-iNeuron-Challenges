# 16. Write a fun which will take input as a dict and give me out as a list of all the values even in case of 2 level nesting it should work

```python
def dict_to_list(d):
    l=[]
    for k,v in d.items():
        if type(v)==dict:
            for m,n in v.items():
                l.append(n)
        else:
            l.append(v)
    return l
```

# 20. write a function by which you will be able to append two PDF files

```python
pip install PyPDF2
```

```python
from PyPDF2 import PdfMerger
def pdf_append(pdfs):
    merger=PdfMerger()
    for pdf in pdfs:
        merger.append(pdf)
    merger.write("result1.pdf")
    merger.close()
    return "Successfully Created result1.pdf"
    
```

```python
l_pdfs=['1.pdf','2.pdf','3.pdf']
pdf_append(l_pdfs)
```
